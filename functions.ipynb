1. What is the difference between a function and a method in Python?

* **Function**: A block of code that performs a task and is defined using `def` outside of a class. It can be called independently.
  Example: `def add(x, y): return x + y`

* **Method**: A function that is associated with an object (usually defined inside a class) and operates on that object.
  Example: `class MyClass: def greet(self): print("Hello")`
  Here, `greet()` is a method.
2.Explain the concept of function arguments and parameters in Python?

* **Parameters** are the variables listed in a function's definition.
  Example: `def greet(name):` → `name` is a parameter.

* **Arguments** are the actual values passed to the function when it is called.
  Example: `greet("Alice")` → `"Alice"` is an argument.

**Parameters** define what a function expects, **arguments** provide the data.
3. What are the different ways to define and call a function in Python?
 - In Python, functions can be defined and called in several ways:

### **1. Regular Function Definition**

```python
def greet(name):
    return f"Hello, {name}"

greet("Alice")  # Function call
```

### **2. Function with Default Arguments**

```python
def greet(name="Guest"):
    return f"Hello, {name}"

greet()         # Uses default
greet("Alice")  # Overrides default
```

### **3. Function with Variable Arguments**

```python
def greet(*args):
    for name in args:
        print(f"Hello, {name}")

greet("Alice", "Bob")
```

### **4. Lambda (Anonymous) Function**

```python
add = lambda x, y: x + y
add(2, 3)  # Returns 5
```

### **5. Function Inside a Function (Nested)**

```python
def outer():
    def inner():
        return "Inner"
    return inner()

outer()  # Calls inner function inside outer
```

### **6. Calling with Keyword Arguments**

```python
def greet(name, age):
    return f"{name} is {age} years old"

greet(age=30, name="Alice")  # Keyword call
```

### **7. Using `*args` and `**kwargs` Together**

```python
def info(*args, **kwargs):
    print(args, kwargs)

info("Python", version=3.10)
```
4.  What is the purpose of the `return` statement in a Python function?
 - The `return` statement in a Python function is used to send a value back to the caller of the function.
5.  What are iterators in Python and how do they differ from iterables?
 - In Python, **iterables** are objects you can loop over (like lists or strings), while **iterators** are objects that keep track of where you are in a loop and produce items one at a time using `__next__()`. An iterator is created from an iterable using the `iter()` function.

6. . Explain the concept of generators in Python and how they are defined.
 - Generators in Python are a type of iterator that yield items one at a time using the `yield` keyword, allowing for memory-efficient iteration. They are defined like regular functions but use `yield` instead of `return`.
7. What are the advantages of using generators over regular functions?
 - Generators are more memory-efficient than regular functions because they yield items one at a time instead of storing the entire result in memory. They also support lazy evaluation, which makes them ideal for working with large data sets or infinite sequences.
8.  What is a lambda function in Python and when is it typically used?
 - A **lambda function** in Python is an anonymous, single-expression function defined using the `lambda` keyword. It’s typically used for short, throwaway functions, especially as arguments to functions like `map()`, `filter()`, or `sorted()`.
9. . Explain the purpose and usage of the `map()` function in Python.
 - The `map()` function in Python applies a given function to each item of an iterable and returns a map object (an iterator). It's used to transform data efficiently without writing explicit loops.
10. . What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - * **`map()`** applies a function to each item in an iterable and returns the results.
* **`filter()`** selects items from an iterable that meet a condition (function returns `True`).
* **`reduce()`** (from `functools`) applies a function cumulatively to reduce the iterable to a single value.
